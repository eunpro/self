```python
import numpy as np
import matplotlib.pylab as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))    
```

```python
import numpy as np
import matplotlib.pylab as plt


def relu(x):
    return np.maximum(0, x) #0과 x중 더 큰 값 취함

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt


def relu(x):
    return np.maximum(0, x) #0과 x중 더 큰 값 취함

x = np.arange(-10.0, 5.0, 0.4)
y = relu(x)
plt.plot(x, y)
plt.xlim(-15,10)
plt.ylim(-1.0, 5.5)
plt.show()
```
